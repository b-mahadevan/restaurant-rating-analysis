# Problem Faced - Encoding Issue: Corrupted City Names

Certain city names containing special characters **(such as í, ã)** appeared corrupted due to improper text encoding during file import or processing.

These issues can lead to incorrect filtering, grouping, and visualizations.

# Solution
Replaced the corrupted entries using a dictionary mapping.

```python
# Define a mapping of corrupted to correct city names

fix_city_name ={
    'Bras�_lia': 'Brasília',
    'S��o Paulo': 'São Paulo',
    '��stanbul': 'Istanbul'
}

#Replace the corrupted city names

df['City'] = df['City'].replace(fix_city_name)
```
